After using the regression and regression line best of fit between assessed value and sales amount, I found a weak, positive relationship between assessed value and sale amount. I discovered that the (r) value, or coefficient, is 0.121608. Instead of using all the real estate data for all the types of homes and residential types, I chose to use only single-family homes because that's where most people looking for some residential property buy. It also eliminates any outliers that could give us an inaccurate representation of the relationships between assessed value and sale amount. I looked at the scatter plot and noticed that most of the single-family home data were mainly clustered around between 0.1 -0.5, with a few outliers. After making a regression line, I noticed that most of the points on the line are in the cluster, while the other half are either above or below the line, which does not conform to the regression line. The answer to my question is that the assessed value is unrelated to the sale amount, which means asset value is not used to predict the sale amount on single-family homes.
