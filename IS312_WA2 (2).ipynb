#**Dataset:**
We can view team-level statistics from recent NBA games, with each row representing a single team's performance in a specific matchup. Each row will include whether the team played at home (home=1) or (away=0) and whether they won the game or not. The key performance metrics that will be used in this analysis are team score, assists, reboundsTotal, pointsFastBreak, pointsFromTurnovers, pointsInThePaint, and pointsSecondChance for both home and away team performance during a matchup. Since each team's metrics correspond to a single team in a single game, this structure makes it ideal for examining whether location affects a team's performance.

#**Data Dictionary:**

1. teamScore: total points the team scored during that one game

2. assists: the number of assists the team had during that one game

3. reboundsTotal: total number of offensive and defensive rebounds collected by the team

4. pointsFastBreak: the total number of fast break points the team scored during the game.

5. pointsFromTurnovers: total number of turnovers leading to points the team had during the game

6. pointsInThePaint: total number of points in the paint during the game

7. pointsSecondChance: total number of team second-chance points during the game.




#**Do Team's Perform Better at Home?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TeamStatistics.csv')
df.head(10)
```

```python
home_games = df[df['home'] == 1]
home_games
```

```python
away_games = df[df['home'] == 0]
away_games
```

```python
win_sums = df.groupby('home')['win'].sum()
game_counts = df.groupby('home')['win'].count()

win_stats = pd.DataFrame({
    'wins': win_sums,
    'total_games': game_counts
})

win_stats.index = ['Away', 'Home']


win_stats['win_percentage'] = (win_stats['wins'] / win_stats['total_games']) * 100
print(win_stats.round(2))

```

According to this dataset covering NBA games from 1972 to the present day, home teams have a 61.69% win rate at home, while away teams have a 38.31% chance of winning. This is almost double the chance of winning at home than away games. With both teams playing the same number of total games (71879), this disparity is not a simple imbalance; it indicates that playing at home increases a team's chances of winning.


```python
#Chatgpt 4.0 prompt 1
from sklearn.linear_model import LinearRegression
import numpy as np


X = df[['home']]
y = df['teamScore']

model = LinearRegression()
model.fit(X, y)


coef = model.coef_[0]
intercept = model.intercept_

print(f"Intercept: {intercept:.2f}")
print(f"Coefficient for home: {coef:.2f}")
```

```python
# chatgpt 4.0 prompt 3
import matplotlib.pyplot as plt
import numpy as np


x_vals = np.array([0, 1])
y_vals = intercept + coef * x_vals


plt.figure(figsize=(6, 4))
plt.plot(x_vals, y_vals, marker='o', linestyle='--', color='green')
plt.xticks([0, 1], ['Away', 'Home'])
plt.title('Predicted Team Score: Home vs Away')
plt.xlabel('Game Location')
plt.ylabel('Predicted Team Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

To quantify the impact of location on the scoring of both home and away teams in the NBA, I created a simple linear regression model where the dependent variable was the total points scored by the team (team score), and the independent variable was the game location (home = 1 and away = 0). The model produced an intercept of 102.27, representing the average score away teams tend to score. The coefficient for the games was +3.58, indicating that playing at home increases a team's expected scoring ability by approximately 3.58 points. This supports the earlier findings of team metrics and in-game-controlled metrics, which indicate that home teams tend to have a slight advantage.

```python
#Chatgpt 4.0 prompt 2
from sklearn.linear_model import LogisticRegression


X = df[['home']]
y = df['win']


log_model = LogisticRegression()
log_model.fit(X, y)

odds_ratio = np.exp(log_model.coef_[0][0])
print(f"Odds of winning at home vs away: {odds_ratio:.2f}")
```

```python
#chatgpt 4.0 prompt 4
home_preds = log_model.predict_proba([[1]])[0][1]
away_preds = log_model.predict_proba([[0]])[0][1]


locations = ['Away', 'Home']
probs = [away_preds, home_preds]

plt.figure(figsize=(6, 4))
plt.bar(locations, probs, color=['gray', 'blue'])
plt.title('Predicted Probability of Winning: Home vs Away')
plt.ylabel('Win Probability')
plt.show()
```

To further investigate the impact of game location on NBA game outcomes, I applied a logistic regression model where the binary response variable was a win (1 = win and 0 = loss), and the predictor variable was home or away (home = 1 and away = 0). The model outputs an odds ratio of 2.59, indicating that the teams are over twice as likely to win when playing at home compared to when playing away. We note the same statistic for the total games played at home vs. away and the win percentage.

The causes of home-court advantage may not be immediately apparent in the numbers alone. Still, the data strongly suggest that being at home correlates with improved in-game metrics across all key basketball metrics. Teams have twice as much chance to win at home as they do away.
